### Plotting the data

#Histogram

```python
col=['2001','2002','2003','2004','2005','2006','2007','2008','2009','2010',
    '2011','2012','2013','2014','2015','2016','2017','2018','2019','2020','2021','2022']
```

```python
for i in col:
    fig=plt.figure(figsize=(5,5))
    plt.hist(df[i],bins=10)
    plt.xlabel(i)
    plt.show()
```

```python
i=1
while i<=22:
    sns.histplot(data=df,bins=10)
    i=i+1
```

### Barplot

```python
# Extract the years from dataframe
years=df.columns[1:]
```

```python
#sum the values for each year across all countries
total_values=df[years].sum()
total_values
```

```python
# plotting the bar chart
plt.figure(figsize=(20,20))
plt.barh(years,total_values,color='green')
plt.xlabel('Total Values',size=20)
plt.ylabel('Year',size=20)
plt.title('Total values per year',size=20)
plt.show()
```

## Insights

```python
# Plotting indicators of top 10 countries for the different years by the base_year 2001
```

```python
country_by_2001=df.sort_values(by='2001').head(10)
country_by_2001
```

```python
country_by_2001_t=country_by_2001.set_index('Country Name').T
```

```python
country_by_2001_t
```

```python
for country_name,data_values in country_by_2001_t.iterrows():
    fig=plt.figure(figsize=(10,5))
    sns.barplot(x=data_values.index,y=data_values.values)
    plt.xlabel('Countries')
    plt.ylabel('Data Values')
    plt.title(f"{country_name} - Data Values from 2001 to 2022")
    plt.xticks(rotation=90)
    plt.show()
```
